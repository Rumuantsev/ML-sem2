1. Создайте объект Series, значениями которого являются целые числа от 1 до 5, а
индексами символы 'a', 'b', 'c', 'd', 'e'.
1. С помощью обращения по явному индексу получите значение 4.
2. С помощью обращения по неявному индексу получите значение 2.
3. Добавьте в серию новый элемент.
4. Используя операцию среза, получите значения 3, 4, 5.
5. Создайте объект DataFrame из списка [[1, 2], [5, 3], [3.7, 4.8]] с именами столбцов
col1 и col2.
6. Используя операцию индексации, получите элемент 3.7.
7. Измените элемент 3 на 9.
8. Используя операцию среза, получите строки с индексами 1 и 2.
9. Добавьте столбец col3, значения которого – результат поэлементного
перемножения col1 и col2

```python
import pandas as pd

s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print("Series:\n", s, "\n")

print("Значение по индексу 'd':", s['d'])

print("Значение по индексу 1:", s.iloc[1])

s['f'] = 6
print("После добавления нового элемента:\n", s, "\n")

print("Срез (3, 4, 5):\n", s['c':'e'], "\n")

df = pd.DataFrame([[1, 2], [5, 3], [3.7, 4.8]], columns=['col1', 'col2'])
print("DataFrame:\n", df, "\n")

print("Элемент 3.7:", df.loc[2, 'col1'])

df.loc[1, 'col2'] = 9
print("После изменения 3 → 9:\n", df, "\n")

print("Срез строк 1 и 2:\n", df.loc[1:2], "\n")

df['col3'] = df['col1'] * df['col2']
print("После добавления col3:\n", df)
```

### Посмотрите информацию о наборе данных о преступности в Лос-Анджелесе (https://data.lacity.org/Public-Safety/Crime-Data-from-2010-to-2019/63jg-8b9z). Используя Pandas, выполните следующие действия:
1. Загрузите случайную выборку из этого набора и далее работайте с ней:
https://github.com/slemeshevsky/python-course-ipynb/blob/master/src-pandas/la-crimessample.csv.zip
2. Сколько строк и столбцов в таблице?
3. Каковы названия столбцов? Какие типы данных у столбцов?
4. Сколько в каждом из них уникальных значений?
5. Сколько пропущенных значений?
6. Верно ли, что женщины чаще оказываются жертвами по сравнению с мужчинами?
7. Определите 10 самых распространенных преступлений в Лос-Анджелесе.
Постройте график.
8. От каких преступлений чаще страдают женщины, а от каких мужчины?
9. Люди какого происхождения чаще всего подвергаются преступлениям?
10. Отсортируйте районы, по количеству преступлений. Постройте график,
показывающий самые безопасный и опасный районы.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://github.com/slemeshevsky/python-course-ipynb/raw/master/src-pandas/la-crimes-sample.csv.zip", 
                 compression="zip")

print("Строк и столбцов, соответственно:", df.shape)

print("\nНазвания столбцов:", df.columns.tolist())
print("\nТипы данных:")
print(df.dtypes)

# 4. Количество уникальных значений
print("\nУникальные значения в каждом столбце:")
print(df.nunique())

# 5. Количество пропущенных значений
print("\nПропущенные значения:")
print(df.isna().sum())

# 6. Женщины или мужчины чаще становятся жертвами?
print("\nРаспределение жертв по полу:")
print(df['Victim Sex'].value_counts(dropna=True))

# 7. Топ-10 самых распространённых преступлений
top_crimes = df['Crime Code Description'].value_counts().head(10)
print("\nТоп-10 преступлений:")
print(top_crimes)

plt.figure(figsize=(10,5))
top_crimes.plot(kind='bar')
plt.title("Топ-10 преступлений в Лос-Анджелесе")
plt.ylabel("Количество")
plt.show()

# 8. Преступления по полу
crime_by_sex = df.groupby(['Victim Sex','Crime Code Description']).size().unstack(fill_value=0)
female_top = crime_by_sex.loc['F'].sort_values(ascending=False).head(5)
male_top = crime_by_sex.loc['M'].sort_values(ascending=False).head(5)

print("\nПреступления против женщин:")
print(female_top)
print("\nПреступления против мужчин:")
print(male_top)

# 9. По какому происхождению чаще жертвы?
print("\nЧаще всего жертвы по происхождению:")
print(df['Victim Descent'].value_counts().head(10))

# 10. Районы по количеству преступлений
districts = df['Area Name'].value_counts()
print("\nРайоны по количеству преступлений:")
print(districts)

plt.figure(figsize=(12,6))
districts.plot(kind='bar')
plt.title("Количество преступлений по районам Лос-Анджелеса")
plt.ylabel("Количество")
plt.show()
```
