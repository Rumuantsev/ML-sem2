Аналитически найдите точку минимума (то есть такое значение переменной $x$, при котором функция принимает минимальное значение) у следующих функций:

1. $f(x) = 3x^{2} - 12x + 4$
2. $f(x) = x(x^{2} - 9)$

### Функция 1
$$
f(x) = 3x^{2} - 12x + 4
$$

**1. Первая производная**
$$
f'(x) = 6x - 12
$$

**2. Кандидаты на точку минимума**
$$
6x - 12 = 0 \quad\Rightarrow\quad x = 2
$$

### Функция 2
$$
f(x) = x(x^{2} - 9) = x^{3} - 9x
$$

**1. Первая производная**
$$
f'(x) = 3x^{2} - 9
$$

**2. Кандидаты на точку минимума**
$$
3x^{2} - 9 = 0 \quad\Rightarrow\quad x^{2} = 3 \quad\Rightarrow\quad x = \pm\sqrt{3}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-1, 5, 400)
x2 = np.linspace(-4, 4, 400)

y1 = 3*x1**2 - 12*x1 + 4
y2 = x2**3 - 9*x2

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'b-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.title('f(x) = 3x² - 12x + 4')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(1, 2, 2)
plt.plot(x2, y2, 'r-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.title('f(x) = x³ - 9x')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.tight_layout()
plt.show()
```

1. Точка минимума x = 2
2. Точка минимума x = корень из 3

3. При каком начальном значении $a_0$ градиентный спуск по функции $f(x) = x^{2}$ при прочих равных приведет к не меньшему перемещению «шарика», чем по функции $f(x) = x^{4}$?
   
   Продемонстрируйте значение $a_1$ для обеих функций при этом $a_0$.

```python
import math

def gd_step(grad, a, h):
    return a - h * grad(a)

grad_x2 = lambda x: 2 * x        
grad_x4 = lambda x: 4 * x**3     
h = 0.1

a0 = 2.0
a1_x2 = gd_step(grad_x2, a0, h)
a2_x2 = gd_step(grad_x2, a1_x2, h)

print("Задание 1 — f(x) = x^2, a0 = 2, h = 0.1")
print("a1 =", a1_x2)
print("a2 =", a2_x2)
if (abs(a2_x2) < abs(a0)):
    print("Движется ли к минимуму")

a0 = 2.0
a1_x4 = gd_step(grad_x4, a0, h)

print("\nЗадание 2 — f(x) = x^4, a0 = 2, h = 0.1")
print("a1 =", a1_x4)
if (abs(a1_x4 - 2.0) > abs(a1_x2 - 2.0)):
    print("Больше движение, чем в 1 случае")

print("Задание 3")
for x in (0.001, 1, 0.001):
    step_x2 = gd_step(grad_x2, x, h)
    step_x4 = gd_step(grad_x4, x, h)
    if (step_x2 <= step_x4):
        print(x)
        print(step_x2)
        print(step_x4)
```

Задача 1

Функция:  
$f(x) = 2^{x}$  

Аналитическая производная:  
$f'(x) = 2^{x} \cdot \ln 2$  

В точке $x = 0$:  
$f'(0) = 2^{0} \cdot \ln 2 = 1 \cdot \ln 2 \approx 0.6931$  


```python
def diff1(f, a, delta=0.05):
    return (f(a + delta) - f(a)) / delta

def diff2(f, a, delta=0.05):
    f_m2 = f(a - 2*delta)
    f_m1 = f(a - delta)
    f_p1 = f(a + delta)
    f_p2 = f(a + 2*delta)
    return (f_m2 - 8*f_m1 + 8*f_p1 - f_p2) / (12 * delta)

f = lambda x: 2**x

a = 0  
delta = 0.05

forward_val = diff1(f, a, delta)
five_point_val = diff2(f, a, delta)

print("Производная по первой формуле", forward_val)
print("Производная по второй формуле", five_point_val)
```

```python
f = lambda x: x*(x**2 - 9) 

a0 = -1.0
h = 0.1
delta = 0.05

grad_diff1 = lambda x: diff1(f, x, delta=delta)
a1_d1 = gd_step(grad_diff1, a0, h)
a2_d1 = gd_step(grad_diff1, a1_d1, h)

print("Используя производную по первой формуле")
print(" a1 =", a1_d1)
print(" a2 =", a2_d1)

grad_diff2 = lambda x: diff2(f, x, delta=delta)
a1_d2 = gd_step(grad_diff2, a0, h)
a2_d2 = gd_step(grad_diff2, a1_d2, h)
print("Используя производную по второй формуле")
print(" a1 =", a1_d2)
print(" a2 =", a2_d2)
print("В обоих случаях стремится к локальному минимуму(видим из графика далее)")
```

Реализуйте метод градиентного спуска для функции многих переменных. Пусть дана функция $f(x) = x_1^2 + 10x_2^2$. Начните с точки $a_0 = (10, 1)$ и выполните процедуру градиентного спуска с шагом $h = 0.1$. Выведите на экран значения $a_1$ и $a_2$. Значение движется к минимуму? Выполните 10 итераций и выведите их на экран.

```python
def grad(a):
    x1, x2 = a
    return [2*x1, 20*x2] # Вычисленно аналитически по формуле частной производной

def gd_step(grad_func, a, h):
    g = grad_func(a)
    return [a[0] - h * g[0], a[1] - h * g[1]]

h = 0.1
a0 = [10.0, 1.0]

iters = [a0]
for i in range(10):
    iters.append(gd_step(grad, iters[-1], h))

print("a1 =", tuple(round(v,6) for v in iters[1]))
print("a2 =", tuple(round(v,6) for v in iters[2]))
print()

print("Итерации (a0 .. a10):")
for k, a in enumerate(iters):
    print(f"a{k}: ({a[0]:.2f}, {a[1]:.2f})")
```

По x1 мы действительно движемся к минимуму, а x2 скачет от минуса к плюсу
