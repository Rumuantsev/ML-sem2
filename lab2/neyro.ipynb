## 2. Решение задач

---

### Функция 1
\[
f(x) = 3x^{2} - 12x + 4
\]

**1. Первая производная**
\[
f'(x) = 6x - 12
\]

**2. Критическая точка (решаем \(f'(x)=0\))**
\[
6x - 12 = 0 \quad\Rightarrow\quad x = 2
\]

**3. Вторая производная**
\[
f''(x) = 6
\]

Поскольку \(f''(x)=6>0\), в точке \(x=2\) — **локальный минимум**.

**Ответ:** точка минимума: **\(x = 2\)**.

---

Функция:
\[
f(x) = x(x^{2} - 9) = x^{3} - 9x
\]

**1. Первая производная**
\[
f'(x) = 3x^{2} - 9
\]

**2. Критические точки (решаем \( f'(x) = 0 \))**
\[
3x^{2} - 9 = 0 \quad\Rightarrow\quad x^{2} = 3 \quad\Rightarrow\quad x = \pm\sqrt{3}
\]

**3. Вторая производная**
\[
f''(x) = 6x
\]

**4. Характер критических точек:**

- Для \(x = -\sqrt{3}\):  
\[
f''(-\sqrt{3}) = 6(-\sqrt{3}) = -6\sqrt{3} < 0
\]  
Это **локальный максимум**.  

- Для \(x = \sqrt{3}\):  
\[
f''(\sqrt{3}) = 6\sqrt{3} > 0
\]  
Это **локальный минимум**.  

**Ответ:** точка минимума:  
\[
x = \sqrt{3}
\]
