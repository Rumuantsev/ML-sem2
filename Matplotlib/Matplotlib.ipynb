#### Создайте рисунок с основными элементами изображения, похожий на приведенный ниже. Элементы изображения обозначены кругами и подписаны синим шрифтом. Допускается изменение линий, цвета, толщины линий и т.п., главное, чтобы все обозначенные элементы присутствовали.



```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator

x = np.linspace(0, 10, 100)
y1, y2, y3 = np.sin(x), np.cos(x), np.sin(x) * 0.5 + 0.5

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y1, 'b-', linewidth=2, label='Линия (график)')
ax.plot(x, y2, 'r--', linewidth=2, label='Другая линия')
ax.scatter(x[::10], y3[::10], color='green', s=50, label='Точечная диаграмма')

ax.xaxis.set_major_locator(MultipleLocator(1))    
ax.xaxis.set_minor_locator(MultipleLocator(0.25))  
ax.tick_params(axis='x', which='minor', labelbottom=False)  

ax.yaxis.set_major_locator(MultipleLocator(0.5))   
ax.yaxis.set_minor_locator(MultipleLocator(0.25))  
ax.tick_params(axis='y', which='minor', labelleft=False)    

ax.grid(True, which='both', linestyle='--', alpha=0.7)

ax.set_xlabel('Подпись оси OX', fontsize=12, fontweight='bold')
ax.set_ylabel('Подпись оси OY', fontsize=12, fontweight='bold')
ax.set_title('Рисунок', fontsize=14, fontweight='bold')
ax.legend(loc='upper right')

plt.show()



```
